```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_missing = pd.read_csv('Wholesale customers data-missing.csv',index_col=0)
data = pd.read_csv('Wholesale customers data.csv',index_col=0)
data_missing.head()
```

```python
# drop all the rows with missing value
data0 = data_missing.dropna(axis=0)
# replace all missing values with -1, and then they can't be used in calculations
data1 = data_missing.fillna(-1)
# replace all missing value with mean
datamean = data_missing.fillna(data_missing.mean().astype(int))
```

```python
# write to files
data0.to_csv(path_or_buf='data0.csv')
data1.to_csv(path_or_buf='data1.csv')
datamean.to_csv(path_or_buf='datamean.csv')
```
