To deal with missing data in threes ways:
drop all rows with missing value,
replace all missing values with -1,
replace all missing value with mean

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_missing = pd.read_csv('Wholesale customers data-missing.csv',index_col=0)
data = pd.read_csv('Wholesale customers data.csv',index_col=0)
data_missing.head()
```

```python
# drop all the rows with missing value
data0 = data_missing.dropna(axis=0)
# replace all missing values with -1, and then they can't be used in calculations
data1 = data_missing.fillna(-1)
# replace all missing value with mean
datamean = data_missing.fillna(data_missing.mean().astype(int))
```

```python
# write to files
data0.to_csv(path_or_buf='data0.csv')
data1.to_csv(path_or_buf='data1.csv')
datamean.to_csv(path_or_buf='datamean.csv')
```

## Display the comparison between the actual values and the means that are used to replace missing values

```python
# show all the rows with missing value
data_missing[data_missing.isnull().any(axis=1)]
```

```python
# show all the means that are used to replace the missing value
dm = datamean[data_missing.isnull().any(axis=1)]
dm
```

```python
# show all the origin data
df = data[data_missing.isnull().any(axis=1)]
df
```

```python
# substract the original values from means
# show how far away they are from the original data
df.subtract(dm, fill_value=0)
```
